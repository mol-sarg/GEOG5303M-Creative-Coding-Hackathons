```python
# Load relevant sheets
stw_df = pd.read_excel(file_path, sheet_name='STW flow and UV')
seabirds_df = pd.read_excel(file_path, sheet_name='Seabird counts (total)')
mst_df = pd.read_excel(file_path, sheet_name='MST Site 1')

# Clean and prepare STW data
stw_df['Date'] = pd.to_datetime(stw_df['Date'], errors='coerce')
stw_df = stw_df[['Date', 'UV Transmittance 1', 'Flow (l/s)']]
stw_df.rename(columns={'UV Transmittance 1': 'UV', 'Flow (l/s)': 'STW Flow'}, inplace=True)

# Clean and prepare seabirds data
seabirds_df['date'] = pd.to_datetime(seabirds_df['date'], errors='coerce')
seabirds_df.rename(columns={'date': 'Date', 'Bird Counts': 'Seabirds'}, inplace=True)
seabirds_df = seabirds_df[['Date', 'Seabirds']]

# Clean and prepare MST data
mst_df.rename(columns={'Sample Date': 'Date', 'Hubac log (10) Gene Copies / 100ml': 'MST Site 1'}, inplace=True)
mst_df['Date'] = pd.to_datetime(mst_df['Date'], errors='coerce')
mst_df = mst_df[['Date', 'MST Site 1']]

# Use uploaded_df and prepare daily averages
merged['DateTime'] = pd.to_datetime(merged['Date'] + ' ' + merged['Time GMT'])
merged.set_index('DateTime', inplace=True)

daily_df = merged.resample('D').mean(numeric_only=True).reset_index()
daily_df.rename(columns={'DateTime': 'Date'}, inplace=True)

# Merge all datasets on Date
merged2 = daily_df.merge(stw_df, on='Date', how='inner')
merged2 = merged2.merge(seabirds_df, on='Date', how='inner')
merged2 = merged2.merge(mst_df, on='Date', how='inner')

# Drop rows with any missing values
merged_clean = merged2.dropna()

# Compute correlation matrix
full_correlation_matrix = merged_clean.corr(numeric_only=True)

# Display result
full_correlation_matrix
```
